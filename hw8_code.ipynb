```python
# Train two SVM models, one with ‘linear’ kernel 
# and another with ‘rbf’ kernel on the dataset (‘label’ as label 
# and all other attributes as features) using 75% of data as training 
# and remaining data as test.
# Calculate accuracy of both models on the test data and report the best accuracy.
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler

data = pd.read_csv('non_linear.csv')
X = data[['X1', 'X2', 'X3']]
y = data['label']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
data.head(5)
```

```python
# pre-processing
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# ‘linear’ kernel 
linear_svm = SVC(kernel='linear')
linear_svm.fit(X_train_scaled, y_train)
# ‘rbf’ kernel
rbf_svm = SVC(kernel='rbf')
rbf_svm.fit(X_train_scaled, y_train)
linear_predictions = linear_svm.predict(X_test_scaled)
rbf_predictions = rbf_svm.predict(X_test_scaled)
linear_accuracy = accuracy_score(y_test, linear_predictions)
rbf_accuracy = accuracy_score(y_test, rbf_predictions)
print(f"accuracy of linear Kernel is: {linear_accuracy:.4f}")
print(f"accuracy of rbf Kernel is: {rbf_accuracy:.4f}")
print(f"the best accuracy is: {max(rbf_accuracy,linear_accuracy):.4f}")
```
