```python
# The dataset contains insurance charges based on the demographic attributes of customers.

# Develop code in a common programming language 
# (Python, R, Java, C#, C++, etc.) to clean the dataset by 
# removing outliers from BMI column
# imputing all missing values
# and normalizing all columns using min-max normalization.

# Submit both the code and the cleaned dataset in Canvas:

# File name of Code : HW1_Code.[file type] (e.g., HW1_Code.py )
# File name of Result : HW1_CleanedDataset.csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

train_data = pd.read_csv('InsuranceCharges.csv')
train_data.head(5)
train_data.info()
train_data.isnull().sum()
```

```python
Q1 = train_data['bmi'].quantile(0.25)
Q3 = train_data['bmi'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = train_data[(train_data['bmi'] < lower_bound) | (train_data['bmi'] > upper_bound)]
# Remove outliers and keep rows with null values in 'bmi'
train_data_no_outliers = train_data[
    (train_data['bmi'] >= lower_bound) & (train_data['bmi'] <= upper_bound) | (train_data['bmi'].isnull())
].copy()
train_data_no_outliers.head(20)
```

```python
# impute missing values in 'bmi' and 'charges'
train_data_no_outliers['bmi'].fillna(train_data_no_outliers['bmi'].median(), inplace=True)
train_data_no_outliers.dropna(subset=['charges'], inplace=True)
train_data_no_outliers.isnull().sum()
```

```python
# normalize all columns using min-max normalization
numeric_cols = train_data_no_outliers.select_dtypes(include=[np.number]).columns.tolist()
for col in numeric_cols:
    min_col = train_data_no_outliers[col].min()
    max_col = train_data_no_outliers[col].max()
    train_data_no_outliers[col] = (train_data_no_outliers[col] - min_col) / (max_col - min_col)
train_data_no_outliers.head(5)
# Save the cleaned dataset
train_data_no_outliers.to_csv('HW1_CleanedDataset.csv', index=False)
```
