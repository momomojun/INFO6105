```python
# Consider weather dataset with ‘RainTomorrow’ 
# as label and other attributes as features.
# Create two RF models and two Adaboost models 
# using 3 and 50 as number of estimators.
# Using a cross-validation method calculate scores of all four models for 5 folds on dataset.
# For each pair of RF and Adaboost models with the same number of estimators compare mean of scores.
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline, make_pipeline
data = pd.read_csv('weather.csv')
print(data.head(5))
data.info()
```

```python
# Consider weather dataset with ‘RainTomorrow’ 
# as label and other attributes as features.
features = ['MinTemp', 'MaxTemp', 'Rainfall', 'WindSpeed', 'Humidity', 'Pressure']
label = ['RainTomorrow']
X = data[features]
y = data[label].values.ravel()
```

```python
# Create two RF models and two Adaboost models 
# using 3 and 50 as number of estimators.
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.metrics import accuracy_score
models = {
    "RandomForest_estimators=3": make_pipeline(
        StandardScaler(),
        RandomForestClassifier(n_estimators=3, random_state=42)
    ),
    "RandomForest_estimators=50": make_pipeline(
        StandardScaler(),
        RandomForestClassifier(n_estimators=50, random_state=42)
    ),
    "AdaBoost_estimators=3": make_pipeline(
        StandardScaler(),
        AdaBoostClassifier(n_estimators=3, random_state=42) 
    ),
    "AdaBoost_estimators=50": make_pipeline(
        StandardScaler(),
        AdaBoostClassifier(n_estimators=50, random_state=42)
    )
}
```

```python
# Using a cross-validation method calculate scores of all four models for 5 folds on dataset.
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.model_selection import cross_val_score
cv_results = {}
for name, model in models.items():
    scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')
    
    cv_results[name] = {
        "scores_list": scores,
        "mean_score": np.mean(scores)
    }
    
    print(f"    {name} score is: {np.round(scores, 4)}")
    print(f"    {name} average score is: {cv_results[name]['mean_score']:.4f}")
    print('\n')
```

```python
# For each pair of RF and Adaboost models with the same number of estimators compare mean of scores.
mean_rf_3 = cv_results["RandomForest_estimators=3"]["mean_score"]
mean_ada_3 = cv_results["AdaBoost_estimators=3"]["mean_score"]

print(f"RF (n_estimators=3) average score: {mean_rf_3:.4f}")
print(f"AdaBoost (n_estimators=3) average score: {mean_ada_3:.4f}")
if mean_rf_3 > mean_ada_3:
    print("RF (n=3) is better")
elif mean_ada_3 > mean_rf_3:
    print("AdaBoost (n=3) is better")
else:
    print("RF is the same score as AdaBoost")

print("\n")

mean_rf_50 = cv_results["RandomForest_estimators=50"]["mean_score"]
mean_ada_50 = cv_results["AdaBoost_estimators=50"]["mean_score"]

print(f"RF (n_estimators=50) average score: {mean_rf_50:.4f}")
print(f"AdaBoost (n_estimators=50) average score: {mean_ada_50:.4f}")
if mean_rf_50 > mean_ada_50:
    print("RF (n=50) is better")
elif mean_ada_50 > mean_rf_50:
    print("AdaBoost (n=50) is better")
else:
    print("RF is the same score as AdaBoost")
```
