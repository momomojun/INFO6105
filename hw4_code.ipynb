```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

#我需要根据之前的血压来输出后面的血压，主要集中点是血压，也就是目标变量
#是否需要数据清洗和数据预处理等操作？
#Train a multiple linear regression on train dataset dataset with "BloodPressure" 
# as the dependent variable (target) and all other features (but not “Outcome”) 
# as independent variables (inputs).
#Using the regression model, predict and set values of ‘BloodPressure’ in  test dataset .
#Train 19 KNN models with k from 1 to 19 on train dataset with "Outcome" as label and all other attributes as features.
#Calculate accuracy of the model for test data (containing predicted values of "BloodPressure").
#Report the K with the highest accuracy
#读文件数据并查看是否成功
#read the data of the file and check if it run
train_data = pd.read_csv('hw4_train.csv')
test_data = pd.read_csv("hw4_test.csv")
train_data.head(5)
```

```python
# train a multiple linear regression
# X is data except BloodPressure and outcome
# Y is BloddPressure
regression_feature = [col for col in train_data.columns if col not in ['BloodPressure', 'Outcome']]
X_train_regression = train_data[regression_feature]
Y_train_regression = train_data['BloodPressure']
#model begin to learn
linear_model = LinearRegression()
linear_model.fit(X_train_regression, Y_train_regression)
# now we have a trained model
```

```python
#predict
#prepare for test data
X_test_predict = test_data[regression_feature]
# use the trained model to predict BloodPressure
predict_BloodPressure = linear_model.predict(X_test_predict)
# create a new data to store predict data
test_data_new = test_data.copy()
test_data_new['BloodPressure'] = predict_BloodPressure
test_data_new.head()
#now we have filled the empty of BloodPressure
test_data_new.to_csv('test_data_new_predict_bloodpressure.csv', index=False)
```

```python
#KNN train
#only predict outcome
KNN_feature = [col for col in train_data.columns if col != "Outcome"]
X_train_KNN = train_data[KNN_feature]
Y_train_KNN = train_data['Outcome']

X_test_KNN = test_data_new[KNN_feature]
Y_test_KNN = test_data_new['Outcome']

#store the accuracy of K
KNN_accuracy = {}
for k in range(1,20):
    #train new KNN model every time
    KNN_model = KNeighborsClassifier(n_neighbors = k)
    KNN_model.fit(X_train_KNN, Y_train_KNN)
    #Y_predict is the data of 'Outcome' we predict
    Y_predict = KNN_model.predict(X_test_KNN)
    #calculate the loss between real data and predict
    accuracy = accuracy_score(Y_test_KNN, Y_predict)
    KNN_accuracy[k] = accuracy
    print(f"k = {k} and k_accuracy = {accuracy:.4f}")

```

```python
K_best = max(KNN_accuracy, key = KNN_accuracy.get)
Best_accuracy = KNN_accuracy[K_best]
print(f"K with the highest accuracy is: {K_best}" + f" The highest accuracy: {Best_accuracy:.3f}")
```
